### Image convertion Coloured image to Grayscale

### User defined path

```python
### Save the image in a specific path by pressing a key from keyboard

import cv2
#image1 = cv2.imread("C:\\Users\\Avik Das\\Documents\\OpenCV\\contents\\betta.jpg",1)  #imread=imageread  
path = input("Enter the path and name of an image : ")  # here path is user defined
print("Your path is : ",path)



#print(image1)

# read image
image1 = cv2.imread(path,0)
#image1 = cv2.imshow("Coloured image",image1)
#image1 = cv2.resize(image1,(1920,1280))   # width , height
print("Give image with grayscale==\n",image1)

# show image
cv2.imshow("Converted Image=",image1)

k=cv2.waitKey(0) #& 0xFF -> if attribute warning arrise this hexicode has to put in 64 bit
if k == ord("s"):
    cv2.imwrite("C:\\Users\\Avik Das\\Documents\\OpenCV\\output.jpg",image1)    # this path is where converted img will save
else:
    cv2.destroyAllWindows()


#cv2.waitKey() # Here parameter inside waitkey handle the life duration of an image ## 5000 milisecond calculation
#cv2.destroyAllWindows()
```
