```python
# Capture Multiple image from a video

import cv2
vidcap = cv2.VideoCapture(0)
ret,image = vidcap.read()  # Read the video
    
    
count = 0

while True:
    if ret == True:
        cv2.imwrite("C:\\Users\\Avik Das\\Documents\\OpenCV\\frames\\imgN%d.jpg" % count,image)  # save frame as JPEG file
        vidcap.set(cv2.CAP_PROP_POS_MSEC,(count**100))  #Used to hold speed of frane generation
        ret,image = vidcap.read()
        cv2.imshow("res",image)
        print('Read a new frame:',count,ret)
        count += 1
        
        if cv2.waitKey(1)  == ord('q'):
            break
            cv2.destroyAllWindows()
            
    else:
        break
        
vidcap.release()
cv2.destroyAllWindows()
```
